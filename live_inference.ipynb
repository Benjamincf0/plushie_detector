```python
from ultralytics import YOLO

import cv2


model_path = 'runs/segment/train7/weights/last.pt'

image_path = 'plushie_yolo2_split/images/val/807d0d84-IMG_8516_Large.jpeg'

img = cv2.imread(image_path)
H, W, _ = img.shape

model = YOLO(model_path, task='segment')  # low thresholds to avoid filtering

results = model(img)

for result in results:
    for j, mask in enumerate(result.masks.data):

        mask = mask.numpy() * 255

        mask = cv2.resize(mask, (W, H))

        cv2.imwrite('./output.png', mask)
```

```python
from ultralytics import YOLO
import time
import cv2

# Load a pretrained YOLO model (recommended for training)
model = YOLO("/Users/benjamin/Code/plushie_detector/runs/segment/train7/weights/last.pt")

cap = cv2.VideoCapture(0)

while cap.isOpened():
    start = time.perf_counter()
    ret, frame = cap.read()
    if not ret:
        break

    # Perform inference on the frame
    results = model(frame, conf=0.9, iou=0.5)

    # Render the results on the frame
    annotated_frame = results[0].plot()

    end = time.perf_counter()
    total_time = end - start
    fps = 1 / total_time

    # Display the annotated frame
    cv2.putText(annotated_frame, f"FPS: {fps:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("YOLOv11 Segmentation", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
